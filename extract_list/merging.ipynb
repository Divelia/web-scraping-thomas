```python
import os
import re
import pandas as pd
from IPython import display
from ipywidgets import interact, widgets

url_origin = 'outputs'
files = os.listdir(url_origin)
print('Total of files founded: ', len(files))
filenames = [re.findall(".*.xlsx", el)[0] for el in files]
print('Files names: ', filenames)
```

```python
# Standard names
bra = ['magalu', 'mundoverde', 'drogariasaopaulo', 'drogaraia', 'lojadosuplemento', 'submarino', 'mercadolibrebr', 'amazonbr', 'americanas', 'natue', 'magazine']
mex = ['mercadolibremx', 'amazonmx', 'walmart', 'sams', 'supernaturista', 'costco', 'sanborns', 'benavides', 'fahorro', 'sanpablofarmacia']
```

```python
list_dataframes = []
for filename in filenames:
    try:
        df = pd.read_excel(url_origin + "//" + filename)
        headers = list(df.columns)
        webpage = re.search(r'.*\-', filename).group(0)[:-1]
        search = re.search(r'\-.*', filename).group(0)[1:-5]
        df['id'] = df.name.str[-2:] + df.image.str[-20:-10]
        df['site'] = webpage
        df['search'] = search
        if webpage in bra:
            df['country'] = 'BRASIL (BRA)'
        elif webpage in mex:
            df['country'] = 'MEXICO (MEX)'
        else:
            df['country'] = 'No encontrado'
        list_dataframes.append(df)
    except:
        continue
```

```python
suma = 0
for i in list_dataframes:
    suma += i.shape[0]
print(suma)
def show_dataframe(n):
    print(list(list_dataframes[n].columns))
    return list_dataframes[n]
opts = [n for n in range(len(list_dataframes))]
interact(show_dataframe, n=widgets.Dropdown(options=opts, value = 0))
```

```python
summary = pd.concat(list_dataframes, ignore_index=True)
print(summary.shape)
summary = summary.dropna(how='all', axis=0).drop_duplicates(keep=False).reset_index(drop=True)
summary.head(10)
```

```python
import re
def extract_quantity(n):
    try:
        y = re.findall(r"(\d+[.,]?\d*)(?:[\s\d\-\+\/]*)(?:Cápsulas|Mg|cápsulas|mL)", n)
        x = y[0]
        if x:
          return x
        else:
          return ''
    except:
        return ''
def extract_brand(n):
    try:
        y = re.findall(r"\s-\s.*", n)
        x = y[0].replace(' -', '')
        if x:
          return x
        else:
          return ''
    except:
        return ''

summary['quantity'] = summary['name'].apply(lambda x: extract_quantity(x))
summary['brand'] = summary['name'].apply(lambda x: extract_brand(x))

summary.head(15)
```
