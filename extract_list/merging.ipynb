```python
# importing packages
import os
import re
import pandas as pd
import numpy as np
from IPython import display
from ipywidgets import interact, widgets
import matplotlib.pyplot as plt
import uuid

# searchs
keywords_table = pd.read_excel('master_table.xlsx', sheet_name="search")
searchs = keywords_table.search.unique().tolist()
searchs = [n.strip() for n in searchs]
print('vitaminas: ', searchs)

# importing data
url_origin = 'outputs'
files = os.listdir(url_origin)
print('Total of files founded: ', len(files))
filenames = [re.findall(".*.xlsx", el)[0] for el in files]
# print('Files names: ', filenames)
```

```python
# Standard names
bra = ['magalu', 'mundoverde', 'drogariasaopaulo', 'drogaraia', 'lojadosuplemento', 'submarino', 'mercadolibrebr', 'amazonbr', 'americanas', 'natue', 'magazine']
mex = ['mercadolibremx', 'amazonmx', 'walmart', 'sams', 'supernaturista', 'costco', 'sanborns', 'benavides', 'fahorro', 'sanpablofarmacia']
```

```python
list_dataframes = []
total_rows = 0
for filename in filenames:
    try:
        df = pd.read_excel(url_origin + "//" + filename)
        headers = list(df.columns)
        webpage = re.search(r'.*\-', filename).group(0)[:-1].strip()
        df['site'] = webpage
        search = re.search(r'\-.*', filename).group(0)[1:-5].strip()
        if search in searchs:
            df['search'] = 'vitaminas'
        else:
            df['search'] = search
        print(webpage, ' has ', df.shape[0], ' datos')
        if webpage in bra:
            df['country'] = 'BRASIL (BRA)'
        elif webpage in mex:
            df['country'] = 'MEXICO (MEX)'
        else:
            df['country'] = 'No encontrado'
        print('columns: ', list(df.columns))
        current_cols = list(df.columns)
        cols_base = ['search', 'site', 'name', 'brand', 'price', 'oldprice', 'quantity', 'image', 'country', 'hyperlink']
        notpresent_cols = list(set(cols_base) - set(current_cols))
        for n in notpresent_cols:
            df[n] = np.nan
        df = df[cols_base]
        list_dataframes.append(df)
        print('\n')
        total_rows += df.shape[0]
    except Exception as e:
        print(e)
        continue
print('TOTAL OF ROWS: ', total_rows)
```

```python
# join all tables
summary = pd.concat(list_dataframes, ignore_index=True)
# deleting rows with nan values
summary = summary.dropna(how='all', axis=0)
summary =  summary[summary['name'].notna()]
summary.head(10)
```

```python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
    )
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s

# normalizing name column to compare and match
summary['name'] = summary['name'].astype(str)
summary['name'] = summary['name'].map(lambda x: x.lower())
summary['name'] = summary['name'].map(lambda x: x.replace("ê", "e"))
summary['name'] = summary['name'].map(normalize)

summary
```

```python
# set list of strings from master table for match
reference_brands = [str(n).lower() for n in master_table['Corporación'].unique().tolist()]
reference_products = [str(n).lower() for n in master_table['Producto'].unique().tolist()]
reference_products_without_spaces = [str(n).lower().replace(' ', '') for n in master_table['Producto'].unique().tolist()]
print('reference_brands: ', reference_brands)
print('reference_products: ', reference_products)

regex_PMLATAMSPA_words = r'(' + '|'.join([n.lower() for n in PMLATAMSPA_words]) + r')'

regex_brands = r'(' + '|'.join(reference_brands) + r')'
exclude_words_regex = r'(' + '|'.join(exclude_words) + r')'

regex_products = r'(' + '|'.join(reference_products) + r')'
regex_products_4_char = r'(' + '|'.join([n[:4] for n in reference_products if len(n)>8]) + r')'
regex_products_5_char = r'(' + '|'.join([n[:5] for n in reference_products if len(n)>9]) + r')'

regex_products_1_char_less = r'(' + '|'.join([n[:-1] for n in reference_products if len(n)>5]) + r')'
regex_products_2_char_less = r'(' + '|'.join([n[:-2] for n in reference_products if len(n)>6]) + r')'
regex_products_3_char_less = r'(' + '|'.join([n[:-3] for n in reference_products if len(n)>7]) + r')'

regex_keywords = r'(' + '|'.join(keywords) + r')'
regex_minerals = r'(' + '|'.join(minerals) + r')'
regex_product_type_list = r'(' + '|'.join([str(n) for n in product_type_list]) + r')'
```

```python
from googletrans import Translator, constants
from pprint import pprint
translator = Translator()
# function to extract matches with quantities type
def extract_quantity_number(n):
    try:
#         query = r"(\d+[.,]?\d*)(?:[\s\d\-\+\/]*)(?:Cápsulas|Mg|cápsulas|capsulas|Cápsulas|mg|mL|tabs)"
        query = r"(\d+)\s\b"
        output = re.search(query, n).group()
        if output:
          return output
        else:
          return ''
    except:
        return ''
    
def extract_quantity_type(n):
    try:
#         query = r"(\d+[.,]?\d*)(?:[\s\d\-\+\/]*)(?:Cápsulas|Mg|cápsulas|capsulas|Cápsulas|mg|mL|tabs)"
        query = r"\s(Cápsulas|Comprimidos|comprimidos|cápsulas|capsulas|Cápsulas|mL|ml|tabs)\b"
        output = re.search(query, n).group()
        if output:
          return output
        else:
          return ''
    except:
        return ''
    
# function to extract matches with quantities type
def extract_content(n):
    try:
        query = r"(\d+)\s(ML|G|MG|IO|mg|NF|K|IU|G|Y|MC|g)\b"
        output = re.search(query, n).group()
        query_1 = r"(\d+)(ML|G|MG|IO|mg|NF|K|IU|G|Y|MC|g)\b"
        output_1 = re.search(query_1, n).group()
        
        if output:
            if output_1:
                return output + ' ' + output_1
            else:
                return output
        else:
          return ''
    except:
        return ''
    
def extract_product_type(n):
    try:
        output = re.search(regex_product_type_list, n).group()
        if output:
            return ' '.join(output)
    except:
        return ''

def extract_kit(n):
    try:
        query = r"(Kit|KIT|kit)\s\d+"
        output = re.search(query, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''
def extract_brand_from_name(n):
    try:
        output = re.search(regex_brands, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''
    
def extract_brand_from_brand(n):
    try:
        output = re.search(regex_brands, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''

        
def extract_official_product_from_name_multiple_conditions(n):
    try:
        output = re.search(regex_products, n).group()
        if output:
          return output
        else:
            output = re.search(regex_products, translator.translate(n)).group()
            if output:
                return output
            else:
                output = re.search(regex_products, translator.translate(n, src='pt', dest='en'))
                if output:
                    return output
                else:
                    output = re.search(regex_products_4_char, n).group()
                    if output:
                        return output
                    else:
                        output = re.search(regex_products_5_char, n).group()
                        if output:
                            return output
                        else:
                            return np.nan
    except:
            return np.nan

def extract_official_product_from_name(n):
    try:
        output = re.search(regex_products, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan
def extract_official_product_from_name_without_spaces(n):
    try:
        output = re.search(reference_products_without_spaces, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan

def extract_official_product_from_name_4_char(n):
    try:
        output = re.search(regex_products_4_char, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan

        
        
def extract_official_product_from_name_1_char_less(n):
    try:
        output = re.search(regex_products_1_char_less, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan
def extract_official_product_from_name_2_char_less(n):
    try:
        output = re.search(regex_products_2_char_less, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan
def extract_official_product_from_name_3_char_less(n):
    try:
        output = re.search(regex_products_3_char_less, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan
    
    
def extract_official_product_from_name_5_char(n):
    try:
        output = re.search(regex_products_5_char, n).group()
        if output:
          return output
        else:
          return np.nan
    except:
            return np.nan
    
        
def extract_keywords(n):
    try:
        output = re.search(regex_keywords, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''
def extract_minerals(n):
    try:
        output = re.search(regex_minerals, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''

        
def extract_excluded_words(n):
    try:
        output = re.search(exclude_words_regex, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''
        
def extract_PMLATAMSPA_words(n):
    try:
        output = re.search(regex_PMLATAMSPA_words, n).group()
        if output:
          return output
        else:
          return ''
    except:
            return ''
        
        
# test_match_no_quantity = test[test['quantity'].isnull()]
test['quantity_number'] = test['name'].apply(lambda x: extract_quantity_number(x))
test['quantity_type'] = test['name'].apply(lambda x: extract_quantity_type(x))
test['quantity_content'] = test['name'].apply(lambda x: extract_content(x))
test['TIPO_DE_PRODUTO_match'] = test['name'].apply(lambda x: extract_product_type(x))
test['kit'] = test['name'].apply(lambda x: extract_kit(x))
test['CORPORACION_name_match'] = test['name'].apply(lambda x: extract_brand_from_name(x))
test['CORPORACION_brand_match'] = test['brand'].apply(lambda x: extract_brand_from_brand(x))
test['KEYWORD_match'] = test['name'].apply(lambda x: extract_keywords(x))
test['MINERALS_match'] = test['name'].apply(lambda x: extract_minerals(x))

test['PRODUCTO_name_match'] = test['name'].apply(lambda x: extract_official_product_from_name(x))
test['PRODUCTO_name_match_without_spaces'] = test['name'].apply(lambda x: extract_official_product_from_name_without_spaces(x))

test['PRODUCTO_name_match_4_char'] = test['name'].apply(lambda x: extract_official_product_from_name_4_char(x))
test['PRODUCTO_name_match_5_char'] = test['name'].apply(lambda x: extract_official_product_from_name_5_char(x))
test['PRODUCTO_name_match_multiple'] = test['name'].apply(lambda x: extract_official_product_from_name_multiple_conditions(x))
test['PRODUCTO_name_match_1_char_less'] = test['name'].apply(lambda x: extract_official_product_from_name_1_char_less(x))
test['PRODUCTO_name_match_2_char_less'] = test['name'].apply(lambda x: extract_official_product_from_name_2_char_less(x))
# test['PRODUCTO_name_match_3_char_less'] = test['name'].apply(lambda x: extract_official_product_from_name_3_char_less(x))
test['words_excluded_found'] = test['name'].apply(lambda x: extract_excluded_words(x))
test['extract_PMLATAMSPA_Producto'] = test['name'].apply(lambda x: extract_PMLATAMSPA_words(x))

test
```

```python
import re

def convert_braz_currency_final_price(c, p):
    if c == 'BRASIL (BRA)':
        try:
            p = p.split('Por', 1)[1]
        except:
            pass
        try:
            y = '.'.join(re.findall('[0-9]+', str(p)))
        except:
            y = p
        if y.count('.') >= 2:
            y = y.replace('.', '', 1)
    else:
        try:
            y = '.'.join(re.findall('[0-9]+', str(p).replace(',', '')))
        except:
            y = str(p).replace(',', '')

    if y:
        return float(y)
    else:
        return np.nan

def convert_braz_currency_full_price(c, p):
    if c == 'BRASIL (BRA)':
        try:
            p = p.split('Por', 1)[0]
        except:
            pass
        try:
            y = '.'.join(re.findall('[0-9]+', str(p).replace('.', '')))
        except:
            y = p
    else:
        try:
            y = '.'.join(re.findall('[0-9]+', str(p).replace(',', '')))
        except:
            y = str(p).replace(',', '')
    
    if y:
        return float(y)
    else:
        return np.nan
    
df['Final_price_customized'] = df.apply(
         lambda row: convert_braz_currency_final_price(row['country'], row['Final Price']),
         axis=1)

df['Full_price_customized'] = df.apply(
         lambda row: convert_braz_currency_full_price(row['country'], row['Full Price']),
         axis=1)

```

```python
df['Final_price_customized'] = df['Final_price_customized'].astype(float)
df['Full_price_customized'] = df['Full_price_customized'].astype(float)
```
