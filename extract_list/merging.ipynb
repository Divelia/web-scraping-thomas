```python
# importing packages
import os
import re
import pandas as pd
import numpy as np
from IPython import display
from ipywidgets import interact, widgets
import matplotlib.pyplot as plt

# importing data
url_origin = 'outputs'
files = os.listdir(url_origin)
print('Total of files founded: ', len(files))
filenames = [re.findall(".*.xlsx", el)[0] for el in files]
# print('Files names: ', filenames)
```

```python
# Standard names
bra = ['magalu', 'mundoverde', 'drogariasaopaulo', 'drogaraia', 'lojadosuplemento', 'submarino', 'mercadolibrebr', 'amazonbr', 'americanas', 'natue', 'magazine']
mex = ['mercadolibremx', 'amazonmx', 'walmart', 'sams', 'supernaturista', 'costco', 'sanborns', 'benavides', 'fahorro', 'sanpablofarmacia']
```

```python
list_dataframes = []
total_rows = 0
for filename in filenames:
    try:
        df = pd.read_excel(url_origin + "//" + filename)
        headers = list(df.columns)
        webpage = re.search(r'.*\-', filename).group(0)[:-1].replace(' ', '')
        search = re.search(r'\-.*', filename).group(0)[1:-5]
        try:
            df['id'] = df.name.str[-2:] + df.image.str[-20:-10]
        except:
            df['id'] = df.image.str[-20:-10]
    
        df['site'] = webpage
        df['search'] = search
        print(webpage, ' has ', df.shape[0], ' datos')
        if webpage in bra:
            df['country'] = 'BRASIL (BRA)'
        elif webpage in mex:
            df['country'] = 'MEXICO (MEX)'
        else:
            df['country'] = 'No encontrado'
        print('columns: ', list(df.columns))
        current_cols = list(df.columns)
        cols_base = ['search', 'site', 'id', 'name', 'brand', 'price', 'oldprice', 'quantity', 'image', 'country']
        notpresent_cols = list(set(cols_base) - set(current_cols))
        for n in notpresent_cols:
            df[n] = np.nan
        df = df[cols_base]
        list_dataframes.append(df)
        print('\n')
        total_rows += df.shape[0]
    except Exception as e:
        print(e)
        continue
print('TOTAL OF ROWS: ', total_rows)
```

```python
# join all tables
summary = pd.concat(list_dataframes, ignore_index=True)

# deleting rows with nan values
summary = summary.dropna(how='all', axis=0)

summary.head(10)
```

```python
# deleting duplicates
summary = summary.drop_duplicates().reset_index(drop=True)
# sending data to excel sheet
summary.to_excel('merge.xlsx', index=False)
summary.head(25)
```

```python
# visualize dataset individually
def show_dataframe(site):
    df = summary[summary['site']==site]
    print('datos: ', df.shape[0])
    print(df.info())
    return df
opts = list(summary.site.unique())
interact(show_dataframe, site=widgets.Dropdown(options=opts, value = 'amazonmx'))
```

```python
reference_brands = master_table['Corporación'].unique().tolist()
reference_products = master_table['Producto'].unique().tolist()
print(summary.shape[0])

# initializing columns
summary.loc[n, 'officialbrand_name_match'] = np.nan
summary.loc[n, 'officialproduct_name_match'] = np.nan
summary.loc[n, 'officialbrand_brand_match'] = np.nan
summary.loc[n, 'minerals_match'] = np.nan
summary.loc[n, 'keyword_match'] = np.nan

print(summary.columns)
for n in range(summary.shape[0]):

    match_officialbrand_name = []
    match_officialproduct_name = []
    match_officialbrand_brand = []
    match_keyword = []
    match_minerals = []
    
    try:
        for official_brand in reference_brands:
            if official_brand in str(summary.loc[n, 'name']).split():
                match_officialbrand_name.append(official_brand)
                
        for official_brand in reference_brands:
            if official_brand in str(summary.loc[n, 'brand']).split():
                match_officialbrand_brand.append(official_brand)
    
        for official_product in reference_products:
            if official_product in str(summary.loc[n, 'name']).split():
                match_officialproduct_name.append(official_product)
                
        for kw in keywords:
            if kw in str(summary.loc[n, 'name']).split():
                match_keyword.append(kw)
                
        for mine in minerals:
            if mine in str(summary.loc[n, 'name']).split():
                match_minerals.append(mine)

    except Exception as e:
        print(e)
    
    if match_officialbrand_name:
        summary.loc[n, 'officialbrand_name_match'] = ', '.join(match_officialbrand_name)
    else:
        summary.loc[n, 'officialbrand_name_match'] = np.nan

    if match_officialproduct_name:
        summary.loc[n, 'officialproduct_name_match'] = ', '.join(match_officialproduct_name)
    else:
        summary.loc[n, 'officialproduct_name_match'] = np.nan
        
    if match_officialbrand_brand:
        summary.loc[n, 'officialbrand_brand_match'] = ', '.join(match_officialbrand_brand)
    else:
        summary.loc[n, 'officialbrand_brand_match'] = np.nan
        
    if match_keyword:
        summary.loc[n, 'keyword_match'] = ', '.join(match_keyword)
    else:
        summary.loc[n, 'keyword_match'] = np.nan
        
    if match_minerals:
        summary.loc[n, 'minerals_match'] = ', '.join(match_minerals)
    else:
        summary.loc[n, 'minerals_match'] = np.nan
```

```python
# function to extract matches with quantities type
def extract_quantity_number(n):
    try:
#         query = r"(\d+[.,]?\d*)(?:[\s\d\-\+\/]*)(?:Cápsulas|Mg|cápsulas|capsulas|Cápsulas|mg|mL|tabs)"
        query = r"(\d+)\s\b"
        output = re.search(query, n).group()
        if output:
          return output
        else:
          return ''
    except:
        return ''
    
def extract_quantity_type(n):
    try:
#         query = r"(\d+[.,]?\d*)(?:[\s\d\-\+\/]*)(?:Cápsulas|Mg|cápsulas|capsulas|Cápsulas|mg|mL|tabs)"
        query = r"\s(Cápsulas|Comprimidos|comprimidos|cápsulas|capsulas|Cápsulas|mL|ml|tabs)\b"
        output = re.search(query, n).group()
        if output:
          return output
        else:
          return ''
    except:
        return ''
    
# function to extract matches with quantities type
def extract_content(n):
    try:
        query = r"(\d+)\s(ML|G|MG|IO|mg|NF|K|IU|G|Y|MC|g)\b"
        output = re.search(query, n).group()
        query_1 = r"(\d+)(ML|G|MG|IO|mg|NF|K|IU|G|Y|MC|g)\b"
        output_1 = re.search(query_1, n).group()
        
        if output:
            if output_1:
                return output + ' ' + output_1
            else:
                return output
        else:
          return ''
    except:
        return ''
    
def extract_product_type(n):
    try:
        output = re.findall(r"(?=("+'|'.join(product_type_list)+r"))", n)
        if output:
            return ' '.join(output)
    except:
        return ''

# test_match_no_quantity = test[test['quantity'].isnull()]
test['quantity_number'] = test['name'].apply(lambda x: extract_quantity_number(x))
test['quantity_type'] = test['name'].apply(lambda x: extract_quantity_type(x))
test['quantity_content'] = test['name'].apply(lambda x: extract_content(x))
test['product_type_match'] = test['name'].apply(lambda x: extract_product_type(x))
test
```

```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup


s = HTMLSession()

url = 'https://www.amazon.com.br/s?k=vitaminas'

def getdata(url):
    r = s.get(url)
    r.html.render(sleep=1)
    soup = BeautifulSoup(r.html.html, 'html.parser')
    return soup

def getnextpage(soup):
    # this will return the next page URL
    pages = soup.find('span', {'class': 's-pagination-strip'})
    print('pages: ', pages)
#     if not pages.find('li', {'class': 'a-disabled a-last'}):
    url = 'https://www.amazon.com.br' + str(pages.find('a', {'class': 's-pagination-item s-pagination-next s-pagination-button s-pagination-separator'}).find('a')['href'])
    print(url)
#         return url
#     else:
#         return


while True:
    data = getdata(url)
    url = getnextpage(data)
    if not url:
        break
    print(url)
```
